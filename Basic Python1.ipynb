```python
import requests
import pandas as pd
```

```python
def get_historic_price(symbol, after='2018-09-01'):

    url = 'https://api.kraken.com/0/public/OHLC'
    pair = f"{symbol.upper()}USD" # XBTUSD when symbol='xbt' for example

    resp = requests.get(url, params={
        "pair": pair,
        'interval': 60,
        'since': str(int(pd.Timestamp(after).timestamp()))
    })
    resp.raise_for_status()

    data = resp.json()

    results_key = [k for k in data['result'].keys() if k != 'last'][0]
    results = [
        (close_time, float(open), float(high), float(low), float(close), float(volume))
        for (close_time, open, high, low, close, vwap, volume, count)
        in data['result'][results_key]
    ]
    df = pd.DataFrame(results, columns=[
        'CloseTime', 'OpenPrice', 'HighPrice', 'LowPrice', 'ClosePrice', 'Volume'
    ])
    df['CloseTime'] = pd.to_datetime(df['CloseTime'], unit='s')
    df.set_index('CloseTime', inplace=True)
    return df
```

```python
# the efficient coding Rec python
print('Welcome to Product Sales Report\n')

# Initialize variables
products = []
total_sales = 0

# Read data and calculate total sales
with open('productReport.txt', 'r') as product_report:
    for line in product_report:
        item_data = line.split()
        products.append(item_data)
        total_sales += int(item_data[2])

print(f'\nTotal Product Sales: {total_sales}\n')

# Display product sales ratios
print('Product Sales Ratios:')
for item_data in products:
    product_percent = int(item_data[2]) / total_sales * 100
    print(f'{item_data[1]}: {product_percent:.2f}%')

# Identify underperforming products
print()
try:
    benchmark = float(input('Enter the percentage threshold for underperforming products: '))
    print('\nUnderperforming Products:')
    for item_data in products:
        product_percent = int(item_data[2]) / total_sales * 100
        if product_percent < benchmark:
            print(f'{item_data[1]}: {product_percent:.2f}%')
except ValueError:
    print('Invalid input. Please enter a numerical value for the threshold.')
```

```python
dad_sport = {'soccer', 'baseball', 'basketball', 'tennis', 'polo', 'softball'}
mom_sport = {'basketball', 'volleyball', 'tennis', 'polo', 'softball', 'rowing'}
kid_sport = {'soccor', 'tennis', 'swimming'}
```

```python
# Union  ใช้ เคร่ื่องหมาย | และ union
print(dad_sport | mom_sport)
print(dad_sport.union(mom_sport))
print(dad_sport.union(mom_sport, kid_sport))
```

```python
# Intersection ใช้เครื่องหมาย & หรือ intersection()
print(dad_sport & mom_sport)
print(dad_sport.intersection(mom_sport))
print(dad_sport.intersection(mom_sport, kid_sport))
```

```python
# Difference ใช้เครื่องหมาย - หรือ difference()
print(dad_sport - mom_sport)
print(dad_sport.difference(mom_sport))
print(dad_sport.difference(mom_sport, kid_sport))
```

```python
# Symmetric difference ใช้เครื่อหมาย ^ หรือ symmetric_difference()
print(dad_sport ^ mom_sport)
print(dad_sport.symmetric_difference(mom_sport))
print(dad_sport.symmetric_difference(mom_sport, kid_sport))
```

```python
# Arbitrary keyword arguments (**kwargs)
# รับ argument ในรูปแบบ dictionary
def greeting (fname, **kwargs):
  title = lname = nname = ''
  for k, v in kwargs.items():
    if k == 'gender' and v == 'male': title = 'Mr.'
    elif k == 'gender' and v == 'female': title = 'Mrs.'
    if k == 'last_name': lname = v
    if k == 'nick_name': nname = ' ('+v+')'
  return 'Hello' + title + fname + ' ' + lname + nname
```

```python
def read_data(file_name):
  with open(file_name, 'r', encoding='utf-8-sig') as f:
    data = f.readlines()
  return data
```

```python
# A function to transform data to the desingnated data structure for you task
def transform_data (data,keys):
  cases = list()
  for i in range(len(data)):
    values = data[i].strip().split(',')
    if len(values) == len(keys):
      case = dict(zip(keys,values))
      cases.append(case)
  return cases
```

```python
# เขียน function ให้ return จำนวนผู้ติดเชื้อ ในวันที่กำหนด num_of_cases
#def num_of_cases(cases, date):
#  n = 0
#  for i in range(len(cases)):
#    if cases[i]['notification_date'] == date:
#      n += 1
#  return n
```

```python
# เขียน function ที่ 2 ให้ return จำนวนผู้ติดเชื้อ ในวันที่กำหนด num_of_cases
def num_of_cases(cases, date):
  n = 0
  for case in cases:
    # Use .get() to avoid KeyError if 'notification_date' is missing
    if case.get('notification_date') == date:
      n += 1
  return n
```

```python
# Main program
data = read_data('/content/drive/MyDrive/Files/confirmed-cases-since-120864.csv')
#print(type(data))
#print(len(data))
# สกัดหัวตาราง เพื่อมาใช้เป็น Key ของ Dictionary
keys = data[0].strip().split(',')
#แปลงข้อมูลให้อยูีในรูปแบบโครงสร้างช้อมูลที่ต้องการ
cases = transform_data(data[1:], keys)
#print(type(cases))
#print(len(cases))
#print(cases[0])

# ดูยอดผู้ติดเชือในวันที่กำหนด
print(num_of_cases(cases, '15/8/2021'))
```
